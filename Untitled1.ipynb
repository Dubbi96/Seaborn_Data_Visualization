```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib as mpl
# 한글폰트 깨지지 않도록
plt.rc('font', family='NanumGothic')
print(plt.rcParams['font.family'])
mpl.rcParams['axes.unicode_minus'] = False

filepath = "./dataset/2019k.csv"

roadkill_data = pd.read_csv(filepath)

roadkill_data.head()
```

```python
f, ax1 = plt.subplots(1, 1, figsize=(17, 10), sharex=True)
sns.countplot(x="Year", data = roadkill_data , color = "gray", ax =ax1)
ax1.set(ylabel='Count',title="Road Kill Count 2004 - 2019")
```

```python
f, ax1 = plt.subplots(1, 1, figsize=(35, 5), sharex=True)
sns.countplot(x="Class", data = roadkill_data)
ax1.set(ylabel='Count',title="Road Kill Species Count 2004 - 2019")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
# 한글폰트 깨지지 않도록
plt.rc('font', family='NanumGothic')
print(plt.rcParams['font.family'])
mpl.rcParams['axes.unicode_minus'] = False

filepath_tot = "./total.csv"
filepath_child = "./children.csv"
tot_data = pd.read_csv(filepath_tot)
child_data = pd.read_csv(filepath_child)
```
